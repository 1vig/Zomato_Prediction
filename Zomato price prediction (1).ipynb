```python
import pandas as pd
from scipy.stats import chi2_contingency

# Assuming the dataset is already stored in df
# Create a list of columns, excluding the target variable 'approx_cost(for two people)'
col_list = list(df.columns)
col_list.remove('approx_cost(for two people)')

# Create a DataFrame for the features' p-value output
results_df = pd.DataFrame(columns=['Features', 'P-value'])

for col in col_list:
    if df[col].dtype == 'object':
        # Create a contingency table
        dataset_table = pd.crosstab(df[col], df['approx_cost(for two people)'])
        
        # Perform Chi-Square test for independence
        chi2_stat, p_value, dof, expected = chi2_contingency(dataset_table)
        
        # Append the results to the DataFrame
        results_df = pd.concat(
            [results_df, pd.DataFrame([[col, p_value]], columns=['Features', 'P-value'])],
            ignore_index=True
        )

# Display the results
print(results_df)
```

```python
import ipywidgets as widgets

# Define a function to plot the selected feature
def plot_feature(feature):
    plt.figure(figsize=(10,6))
    df[feature].hist(bins=30)
    plt.title(f'Distribution of {feature}')
    plt.xlabel(feature)
    plt.ylabel('Frequency')
    plt.figtext(0.5, 0.01, 'Interactive Feature Plot', ha='center', va='center', fontsize=20)

# Create a dropdown widget with the dataframe's column names
dropdown = widgets.Dropdown(options=df.columns, description='Feature:')

# Use the interact function to create the widget and the plot
widgets.interact(plot_feature, feature=dropdown);
```

```python
def perform_eda(df: pd.DataFrame):
    print("========================================")
    print("Exploratory Data Analysis")
    print("========================================\n")

    # Print the shape of the DataFrame
    print("Shape of the DataFrame:")
    print(f"Number of rows: {df.shape[0]}")
    print(f"Number of columns: {df.shape[1]}\n")

    print("========================================\n")

    # Print the column names
    print("Columns in the DataFrame:")
    for col in df.columns:
        print(col)
    print("\n========================================\n")

    # Print the information about the DataFrame
    print("Information about the DataFrame:")
    df.info()
    print("\n========================================\n")

    # Print the number of null values in each column
    print("Number of null values in each column:")
    print(df.isnull().sum())
    print("\n========================================\n")

    # Print the number of unique values in each column
    print("Number of unique values in each column:")
    print(df.nunique())
    print("\n========================================\n")

    # Print the descriptive statistics for each column
    print("Descriptive statistics for each column:")
    return df.describe()
    print("\n========================================\n")
```
